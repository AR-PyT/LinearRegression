### Task is to find the maximum temperature given the minimum temperature
Start by processing data to get training set with fields relevant to the objective

```python
processed_set = summ_weather[['STA', 'Precip', 'WindGustSpd', 'Snowfall', 'PoorWeather', 'MinTemp', 'MaxTemp']]
```

```python
station_locs['STA'] = station_locs['WBAN']
processed_set = pd.merge(summ_weather, station_locs, on='STA')
processed_set.drop(['STA', 'Date', 'MeanTemp', 'YR', 'MO', 'DA', 'PRCP', 'DR', 'FTI', 'SPD', 'TSHDSBRSGF', 'MAX', 'MIN', 'MEA', 'SNF', 'SND', 'FT', 'FB', 'ITH', 'PGT', 'SD3', 'RHX', 'RHN', 'RVG', 'WTE', 'WBAN', 'NAME', 'STATE/COUNTRY ID', 'LAT', 'LON'], axis=1, inplace=True)
```

```python
process_set.describe()
```

```python
processed_set.drop(['WindGustSpd'], axis=1, inplace=True)
```

```python
processed_set['PoorWeather'] = pd.DataFrame(processed_set['PoorWeather'] == 1, dtype=int)
```

```python
processed_set['Snowfall'] = pd.DataFrame(processed_set['Snowfall'] != 0, dtype=int)
```

```python
def is_not_numeric(x):
    try:
        float(x)
        return False
    except ValueError:
        return True
processed_set.drop(processed_set[processed_set.apply(lambda x: is_not_numeric(x['Precip']), axis=1)].index, inplace=True)
```

```python
processed_set['Precip'] = pd.DataFrame(processed_set['Precip'], dtype=float)
```

```python
processed_set.describe()
```

```python
labels = processed_set['MaxTemp']
train_data = processed_set.drop(['MaxTemp'], axis=1)
```

```python
np.save('ProcessedData/WeatherConditions/labels.npy', labels.to_numpy())
np.save('ProcessedData/WeatherConditions/train_data.npy', train_data.to_numpy())
```
